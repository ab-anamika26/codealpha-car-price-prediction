```python
# ==============================
# STEP 1: Import Required Libraries
# ==============================

# Pandas is used for reading and handling datasets
import pandas as pd

# NumPy is used for numerical calculations
import numpy as np

# Matplotlib is used for creating graphs
import matplotlib.pyplot as plt

# Seaborn is used for beautiful statistical visualizations
import seaborn as sns

# Display graphs directly inside the notebook
%matplotlib inline
```

```python
# Read the CSV file and store it in a DataFrame named df
df = pd.read_csv("car data.csv")
```

```python
# Show the first five rows of the dataset
df.head()
```

```python
# Display the number of rows and columns
df.shape
```

```python
# Show all column names
df.columns
```

```python
# Display information about the dataset
df.info()
```

```python
# Display statistical summary of numerical columns
df.describe()
```

```python
# Count missing values in each column
df.isnull().sum()
```

```python
# Remove all rows containing missing values
df.dropna(inplace=True)
```

```python
# Count duplicate rows
df.duplicated().sum()
```

```python
# Delete duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Display the data type of each column
print(df.dtypes)
```

```python
# Display unique values in each column
for column in df.columns:
    print(f"\nColumn: {column}")
    print(df[column].unique())
```

```python
# Create a histogram of car prices
plt.figure(figsize=(8,5))

plt.hist(df["Selling_Price"], bins=30)

plt.title("Distribution of Car Prices")
plt.xlabel("Price")
plt.ylabel("Number of Cars")

plt.show()
```

```python

# Calculate correlation between numerical columns
correlation = df.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(12,8))

sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

```python
# ==============================
# STEP 5.5: Mileage vs Price
# ==============================

plt.figure(figsize=(8,5))

plt.scatter(df["Driven_kms"], df["Selling_Price"])

plt.title("Mileage vs Price")
plt.xlabel("Driven_kms")
plt.ylabel("Selling_Price")

plt.show()
```

```python
# ==============================
# STEP 6: Feature Engineering
# ==============================

# Create a new column called Car_Age
# Assuming the current year is 2026
df["Car_Age"] = 2026 - df["Year"]

# Display first few rows
df.head()
```

```python
# ==============================
# STEP 7: Remove Year Column
# ==============================

# Remove the Year column
df.drop("Year", axis=1, inplace=True)

# Display first five rows
df.head()
```

```python
# ==============================
# STEP 8: Encode Categorical Columns
# ==============================

from sklearn.preprocessing import LabelEncoder

# Create LabelEncoder object
le = LabelEncoder()

# Convert text columns into numbers
df["Car_Name"] = le.fit_transform(df["Car_Name"])
df["Fuel_Type"] = le.fit_transform(df["Fuel_Type"])
df["Selling_type"] = le.fit_transform(df["Selling_type"])
df["Transmission"] = le.fit_transform(df["Transmission"])

# Display first five rows
df.head()
```

```python
# ==============================
# STEP 9: Feature Selection
# ==============================

# Features (Independent Variables)
X = df.drop("Selling_Price", axis=1)

# Target (Dependent Variable)
y = df["Selling_Price"]

# Display shape
print("Features Shape :", X.shape)
print("Target Shape :", y.shape)
```

```python
# ==============================
# STEP 10: Split the Dataset
# ==============================

# Import train_test_split
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X,                  # Input features
    y,                  # Target variable
    test_size=0.2,      # 20% data for testing
    random_state=42     # Ensures the split is the same every time
)

# Display the shapes of the split datasets
print("Training Features:", X_train.shape)
print("Testing Features :", X_test.shape)
print("Training Target  :", y_train.shape)
print("Testing Target   :", y_test.shape)
```

```python
# ==============================
# STEP 11: Train Linear Regression Model
# ==============================

# Import Linear Regression model
from sklearn.linear_model import LinearRegression

# Create the model
model = LinearRegression()

# Train the model using the training data
model.fit(X_train, y_train)
```

```python
# ==============================
# STEP 12: Predict Selling Prices
# ==============================

# Predict prices for the testing data
y_pred = model.predict(X_test)

# Display first 10 predictions
print(y_pred[:10])
```

```python
# ==============================
# STEP 13.1: Import Evaluation Metrics
# ==============================

# Import performance evaluation metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# ==============================
# STEP 13.2: Mean Absolute Error (MAE)
# ==============================

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Display MAE
print("Mean Absolute Error:", mae)
```

```python
# ==============================
# STEP 13.3: Mean Squared Error (MSE)
# ==============================

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Display MSE
print("Mean Squared Error:", mse)
```

```python
# ==============================
# STEP 13.4: Root Mean Squared Error (RMSE)
# ==============================

# Calculate Root Mean Squared Error
rmse = mean_squared_error(y_test, y_pred) ** 0.5

# Display RMSE
print("Root Mean Squared Error:", rmse)
```

```python
# ==============================
# STEP 13.5: R2 Score
# ==============================

# Calculate R2 Score
r2 = r2_score(y_test, y_pred)

# Display R2 Score
print("R2 Score:", r2)
```

```python
# ==============================
# STEP 14: Actual vs Predicted Prices
# ==============================

# Create a comparison DataFrame
comparison = pd.DataFrame({
    "Actual Price": y_test,
    "Predicted Price": y_pred
})

# Display first 10 rows
comparison.head(10)
```

```python
# ==============================
# STEP 15: Actual vs Predicted Graph
# ==============================

# Set figure size
plt.figure(figsize=(8,6))

# Scatter plot
plt.scatter(y_test, y_pred)

# Graph title
plt.title("Actual Price vs Predicted Price")

# X-axis label
plt.xlabel("Actual Selling Price")

# Y-axis label
plt.ylabel("Predicted Selling Price")

# Show graph
plt.show()
```

```python
# ==============================
# STEP 16: Predict a New Car Price
# ==============================

# Example input
new_car = [[50, 6.5, 35000, 2, 0, 1, 0, 5]]

# Predict selling price
predicted_price = model.predict(new_car)

# Display result
print("Predicted Selling Price:", predicted_price[0])
```

```python
# ==============================
# STEP 17: Save the Trained Model
# ==============================

# Import joblib
import joblib

# Save the trained model
joblib.dump(model, "car_price_prediction_model.pkl")

print("Model saved successfully!")
```

```python
# ==============================
# STEP 18: Load Saved Model
# ==============================

# Load the saved model
loaded_model = joblib.load("car_price_prediction_model.pkl")

print("Model loaded successfully!")
```

```python
# ==============================
# STEP 19: Predict Using Saved Model
# ==============================

prediction = loaded_model.predict(new_car)

print("Predicted Selling Price:", prediction[0])
```
